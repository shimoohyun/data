# **연산자**

## <산술 연산자>

```python
a = 11
b = 4

print('덧셈 결과:', a + b)
print('뺄셈 결과:', a - b)
print('곱셈 결과:', a * b)
print('나눗셈 결과:', a / b)
print('나눗셈 정수 몫:', a // b)
print('나눗셈 나머지:', a % b)
print('거듭제곱 결과:', a ** b)
```

## <관계 연산자>

```python
a = 11
b = 4

print(a > b)
print(a >= b)
print(a < b)
print(a <= b)
```

##<논리 연산자>

```python
a = True
b = False
print(a and b)
print(a or b)
print(not a)
print(not b)
```

## <멤버 연산자>

```python
a = 3
B = [1, 3, 5, 7, 9]

print(a in B)
print(a not in B)
```

#**조건문과** **반복문**

##<조건문>

```python
a = 5

if a == 5:
  print('Right!')
  print('a is 5')

if a == 3:
  print('Right!')
  print('a is 3')

if a != 3:
  print('Right!')
  print('a is not 3')
```

##<다중 조건문 1>

```python
a = 5

if a == 5:
  print('Right!')
  print('a is 5')
else:
  print('a is not 5')

a = 3

if a == 5:
  print('Right!')
  print('a is 5')
else:
  print('a is not 5')
```

##<다중 조건문 2>

```python
a = 5

if a < 5:
  print('a is smaller than 5')
elif a > 5:
  print('a is larger than 5')
else:
  print('a is 5')
```

##<while 반복문>

```python
a = 5
i = 1

print('구구단 5단')

#9번 반복하기
while i <= 9:
  print(str(a) + ' X ' + str(i) + ' = ' + str(i*a))
  i = i + 1
```

##<for 반복문>

```python
a = 5

print('구구단 5단')
for i in range(1,10):
  print(str(a) + ' X ' + str(i) + ' = ' + str(i*a))
```

# 자료형

##<숫자형 데이터>

```python
integer_1 = 3214
integer_2 = -128109
float_1 = -1.986214
float_2 = 123.e2

print(integer_1, type(integer_1))
print(integer_2, type(integer_2))
print(float_1, type(float_1))
print(float_2, type(float_2))

print(integer_1 / integer_2, type(integer_1 / integer_2))
```

##<문자열 데이터>

```python
string_5 = "This is String! \"따옴표\" 기호:!@#$%^"
print(string_5, type(string_5))
```

##<콜렉션 자료형>

###<튜플>

```python
tuple_1 = 1, 2, 3, 4, 5
tuple_2 = ('가', '나', '다', '라', '마')
tuple_3 = '파이썬', 10000, False
tuple_4 = '파이썬', (10000, '만큼','어려워'), False

print(tuple_1, type(tuple_1))
print(tuple_2, type(tuple_2))
print(tuple_3, type(tuple_3))
print(tuple_4, type(tuple_4))

tuple_1[2] = 100
```

###<세트>

```python
set_1 = {1, 2, 3, '가', '나', '다', 1, 2}
set_2 = set({1, 2, 3, '가', '나', '다', 1, 2})
set_3 = set([1, 2, 3, '가', '나', '다', 3])

print(set_1)
print(set_2)
print(set_3)
```

###<세트 변경 및 복제>

```python
set_1 = {1, 2, 3, '가', '나', '다'}

set_1.add('추가')
print(set_1)
set_1.remove('가')
print(set_1)
set_copy_1 = set_1.copy( )
print(set_copy_1)
set_copy_1.clear( )
print(set_copy_1)
```

###<리스트 생성>

```python
list_1 = [1, 2, 3, 4, 5, 1, 3]
list_2 = []
print(list_1)
print(list_2)
print(len(list_1))
```

###<리스트 인덱싱/슬라이싱>

```python
a = [1, 2, 3, 4, 5]
print(a[3])
b = a[1:4]
print(b)
```

###<리스트 데이터 정렬>

```python
list_1 = [897, 2, 1, 4, 99, 5.24, 17]
print(list_1)

#뒤집기
list_1.reverse()
print(list_1)

#오름차순 정렬하기
list_1.sort()
print(list_1)

#내림차순 정렬하기
list_1.sort(reverse=True)
print(list_1)
```

###<딕셔너리>

```python
#딕셔너리 생성하기
dict_1 = {'name': '홍길동', 'birth': 1990, 'addr': 'KR'}
print(dict_1)
print(dict_1['birth'])

#키와 값 추가하기
dict_1['weight'] = 60.5
dict_1['family'] = ['아빠', '엄마', '여동생']
print(dict_1)

#여러 키와 값을 동시에 추가하기
dict_1.update({'weight':67.8,'hobby': ['게임', '독서']})
print(dict_1)

#딕셔너리 값 변경하기
dict_1['hobby'] = ['축구','등산']
print(dict_1)

#데이터 삭제하기
del dict_1['weight']
del dict_1['birth']
del dict_1['addr']
print(dict_1)
```
