```python
import pandas as pd
import numpy as np

# --------------------------------------------------
# Excel 파일 불러오기
# --------------------------------------------------
# - "중심경향치.xlsx" 파일을 불러옴
# - sheet_name=0 : 첫 번째 시트를 불러옴 (Python은 인덱스가 0부터 시작)
# - header=0     : 첫 줄을 변수명으로 인식

url = "https://raw.githubusercontent.com/shimoohyun/data/main/personal_info.xlsx"
data = pd.read_excel(url, sheet_name=0, header = 0)
data.head()
# 업로드할 경우
# data = pd.read_excel("personal_infoxlsx", sheet_name=0, header=0)
```

```python
# 중심 경향치

# 평균 (Mean) - 결측값 제외
print("\n평균(키):", data["키"].mean(skipna=True))

# 중앙값 (Median) - 결측값 제외
print("중앙값(키):", data["키"].median(skipna=True))

# 최빈값 (Mode) - 성별 변수
print("최빈값(성별_남자):", data["성별_남자"].mode()[0])
```

```python
# 산포도

# 범위 (Range) = 최대값 - 최소값
range_value = data["키"].max(skipna=True) - data["키"].min(skipna=True)
print("\n범위(Range):", range_value)
# 분산 (Variance)
print("분산(Variance):", data["키"].var(skipna=True))
# 표준편차 (Standard Deviation)
print("표준편차(Standard Deviation):", data["키"].std(skipna=True))
```

```python
# 사분위수 (Quantiles): Q1(25%), Q2(50%, 중앙값), Q3(75%)
quantiles = data["몸무게"].quantile([0.25, 0.5, 0.75])
print("\n사분위수 (Quantiles):")
print(quantiles)

# IQR (Interquartile Range) = Q3 - Q1
iqr = quantiles[0.75] - quantiles[0.25]
print("\nIQR (Interquartile Range):", iqr)
```

```python
# 데이터 요약: describe()

import pandas as pd

# CSV 불러오기
url = "https://raw.githubusercontent.com/shimoohyun/data/main/employee_health_lifestyle.csv"
data = pd.read_csv(url, encoding="utf-8")

# 혹은 data = pd.read_csv("employee_health_lifestyle.csv", encoding="utf-8")

data['gender'] = data['gender'].astype('category')

# 1. 단일 변수(예: sleep_hours)의 요약 통계량
print("===== 단일 변수 요약 통계량 (sleep_hours) =====")
print(data["sleep_hours"].describe())

# 2. 데이터프레임 전체 변수에 대한 요약 통계량
print("\n===== 전체 변수 요약 통계량 =====")
print(data.describe(include="all"))
```

```python
# 왜도 및 첨도

#왜도
data['stress_level'].skew(skipna=True)

#첨도
data['stress_level'].kurt(skipna=True)

#성별(그룹)에 따른 스트레스의 왜도 비교
data.groupby('gender')['stress_level'].skew(skipna=True)
```

```python
# 빈도, 비율, 백분율

# 성별 빈도
print("성별 빈도:\n", data["gender"].value_counts())

# 성별 비율(상대도수)
print("성별 빈도:\n", data["gender"].value_counts(normalize = True))

# 성별 백분율(%)
print("성별 빈도:\n", (data["gender"].value_counts(normalize = True) * 100).round(1))

```

```python
# 통계량 전체 보는 방법(참고용)
import pandas as pd
# CSV 불러오기
url = "https://raw.githubusercontent.com/shimoohyun/data/main/employee_health_lifestyle.csv"
data = pd.read_csv(url, encoding="utf-8")
# 1. 전체 기술통계 요약
print("===== 기술통계 요약 =====")
print(data.describe(include="all"))
# 2. 특정 변수 요약 (예: stress_level)
print("\n===== 특정 변수 요약 =====")
print("평균:", data["stress_level"].mean(skipna=True))
print("표준편차:", data["stress_level"].std(skipna=True))
print("왜도(Skewness):", data["stress_level"].skew(skipna=True))
print("첨도(Kurtosis):", data["stress_level"].kurt(skipna=True))
print("성별 빈도수:\n", data["gender"].value_counts())
# 3. 그룹별 탐색 (성별별 스트레스 평균)
print("\n===== 성별 그룹별 스트레스 평균 =====")
print(data.groupby("gender")["stress_level"].mean())
```

```python
# 교차표(Crosstab)

# 1. 두 개 범주형 변수의 교차표
print("===== 성별 × 연령그룹 교차표 =====")
ct = pd.crosstab(data["gender"], data["age_group"])
print(ct)
# 2. 행/열 합계 포함된 교차표 (R addmargins과 유사)
print("\n===== 성별 × 연령그룹 교차표 (합계 포함) =====")
ct_with_margins = pd.crosstab(data["gender"], data["age_group"], margins=True, margins_name="Total")
print(ct_with_margins)
```
