##<데이터 기본구조 확인>

```python
import pandas as pd
!wget https://raw.githubusercontent.com/shimoohyun/data/main/crime.csv
data = pd.read_csv('crime.csv', encoding='cp949')

print(data.info()) # 변수명, 자료형, 결측치여부 등 확인
print(data.dtypes) # 각 변수의 자료형만 출력
```

```python
# 행과 열의 수 확인 (dim(data))
print("데이터 크기 (행, 열):", data.shape)
# 행(row) 수, 열(column) 수 (nrow(data), ncol(data))
print("행 개수:", data.shape[0])
print("열 개수:", data.shape[1])
# 열 이름 확인 (colnames(data))
print("열 이름:", data.columns.tolist())
# 행 이름 확인 (rownames(data))
print("행 이름(인덱스):", data.index.tolist())
# 앞부분/뒷부분 일부 확인 (head(data), tail(data))
print("\n데이터 앞부분:")
print(data.head())    # 기본 5행 출력
print("\n데이터 뒷부분:")
print(data.tail())    # 기본 5행 출력
```

##<결측값 제거 및 대체>

```python
import pandas as pd
from sklearn.impute import SimpleImputer, KNNImputer

# 예시 데이터 불러오기
!wget https://raw.githubusercontent.com/shimoohyun/data/main/gender_test_missing.csv
data = pd.read_csv("gender_test_missing.csv", encoding="utf-8")

print("===== 원본 데이터 =====")
print(data)

# -------------------------------
# 결측값이있는 행 삭제
new_data = data.dropna()
print(new_data)

# -------------------------------
# 1. 평균값 대체
data["score_mean"] = data["score"].fillna(data["score"].mean())

# 2. 중앙값 대체
data["score_median"] = data["score"].fillna(data["score"].median())

# 3. 최빈값 대체
data["gender_mode"] = data["gender"].fillna(data["gender"].mode()[0])

# 4. 이전 값으로 대체 (Forward Fill)
data["score_ffill"] = data["score"].fillna(method="ffill")

# 5. 이후 값으로 대체 (Backward Fill)
data["score_bfill"] = data["score"].fillna(method="bfill")

# 6. SimpleImputer (평균값)
imputer_mean = SimpleImputer(strategy="mean")
data["score_sklearn_mean"] = imputer_mean.fit_transform(data[["score"]])

# 7. KNNImputer (3개의 최근접 이웃 활용)
imputer_knn = KNNImputer(n_neighbors=3)
data_knn = imputer_knn.fit_transform(data[["score"]])
data["score_knn"] = data_knn

print("\n===== 결측값 대체 결과 =====")
print(data.head(10))
```
